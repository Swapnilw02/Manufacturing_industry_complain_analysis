```python
df = df.dropna(thresh=len(df.columns) - 10) 
```

```python
print(df.isnull().sum())
```

```python
if df['CAMPAIGN_NBR'].isnull().all():
    df = df.drop(columns=['CAMPAIGN_NBR'])
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for col in numeric_cols:
    median = df[col].median()
    df[col] = df[col].fillna(median)
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
for col in categorical_cols:
    df[col] = df[col].fillna('NOT MENTIONED')
```

```python
print(df.isnull().sum())
```

```python
for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip().str.title() 
```

```python
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
## REMOVING OUTLIERS IN COLUMN KM
if 'KM' in df.columns:
    df = df[df['KM'] < 1_000_000]
```

```python
## REMOVING OUTLIERS IN COLUMN REPAIR_AGE
if 'REPAIR_AGE' in df.columns:
    df = df[(df['REPAIR_AGE'] >= 0) & (df['REPAIR_AGE'] <= 5000)] 
```

```python
## REMOVING OUTLIERS IN COLUMNS COST (# No negative costs)
for cost_col in ['TOTALCOST', 'LBRCOST', 'REPORTING_COST']:
    if cost_col in df.columns:
        df = df[df[cost_col] >= 0]
```

```python
cleaned_file_path = r"C:\D DRIVE\Componies Applications\Axion\DA_Task_2_CLEANED.xlsx"
df.to_excel(cleaned_file_path, index=False)

print(f"\n Cleaned data saved to: {cleaned_file_path}")
```
